```python
import sys

import torch

sys.path.append("..")
import os

modelspath = "../save/ResNet18/models/"
# Iterate over all models in the directory
cons_files = {}
for modelname in os.listdir(modelspath):
    modelpath = os.path.join(modelspath, modelname)
    print(f"Checking file: {modelpath}")
    # The files we want are in format save/ResNet18/models/CE_rho0/RDMConsistency_CE_rho0.pt
    # Look inside modelpath for files that start with RDMConsistency and end with .pt
    # Stop on the first since there should only be one per model
    for filename in os.listdir(modelpath):
        if filename.startswith("RDMConsistency") and filename.endswith(".pt"):
            print(f"Found file: {filename}")
            cons_files[modelname] = torch.load(os.path.join(modelpath, filename))
            break
    else:
        print(f"No RDMConsistency file found in {modelpath}")
        continue

cons_files = dict(sorted(cons_files.items(), key=lambda item: item[0]))
print("Loaded consistency files for models:")
for modelname in cons_files.keys():
    print(modelname)
```
